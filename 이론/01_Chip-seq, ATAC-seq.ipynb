# **ChIP-seq / ATAC-seq**

유전자 발현에 영향을 주는 것

* Transcription Factor의 조절
* DNA의 open/ closed 구조

# **ChIP-seq(Chromatin Immunoprecipitation-sequencing)**

01_Chip-seq.jpg

* 특정 단백질과 결합된 DNA를 면역학적으로 antibody를 이용하여 lprecipitation 시킨 후, DNA를 따로 분리하여 NGS를 이용해 sequencing 한 것
* 죽, 세포 내 특정 단백질과 binding하는 DNA sequence를 알아내는 방법

**Transcription factor의 결합 부위와의 기작을 연구하는 데 이용**

#**ATAC-seq(Assay for Transposase-Accesible Chromatin using sequencing**

01_ATAC-seq.jpg

* Tn5 transposase가 접근 가능한 chromatin을 sequencing하는 방법
 - transposase: transposon이 유전체의 다른 부분으로 이동하는 것을 촉해하는 효소
 - Tn5 transposase: ATAC-seq을 위한 돌연변이
* open되어있는 유전체에 마구잡이로 들어가 adaptor DNA sequence를 붙임과 동시에 잘라내고, 이 adaptor를 이용해 PCR 진행
 - 히스톤 단백질 등에 의해 감싸있거나 화학적 변형을 통해 닫혀있는 유전체 구조에는 transposase가 접근 힘들다
* open chromatin region을 분석하여 accesibility of chromatin이 어디인지 알면 유전자 발현 기작에 대해 분석 가능

**open/closed chromatin을 구별할 수 있는 분자 생물학적 실험 방법**
**유전자 발현에 대한 분석에 사용, 후성 유전학 연구에서 사용**
* open chromatin 구조: 유전자의 전사와 발현이 활발한 지역
* closed chromatin 구조: 유전자의 전사와 발현이 억제된 지역
