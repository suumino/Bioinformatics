#**DNA 분석 파이프라인_01_소개**

* DNA: 생명체의 유전정보를 담고 있는 물질
* 구성요소: AGCT
  - 한 사람의 유전체에는 몇개의 염기서열이 있을까? 30억개
* Central Dogma: DNA -> RNA > Protein
* 표현형(Phenotype)
* DNA를 분석: DNA서열을 읽어 기준 서열과 다른 염기를 찾는 것
* DNA 서열 읽는 법
  - 읽기 -> 자르기 -> sequencer -> 읽힘
* 기준 서열(Reference Sequence): **Genome Reference Consortium** - 기준 서열 제공
* 다른 염기(변이)를 찾는 법: alignment
  - read를 포함한 파일 형태: FASTQ
  - read를 정렬해서 나온 것: BAM
  - 변이를 table 형태로 만든 것: VCF

**이러한 과정을 변이를 찾는 파이프라인 이라고 한다**

* DNA 찾는 일련의 과정을 표준화 시킨 것: GATK Best Practice

#**DNA 분석 파이프라인_02_인트로**

* 강의 목표
  - DNA -> Sequencer -> Raw data -> Analysis -> Find Variant
  - public raw data data에서 나타난 유전적 변이(variant)들을 생물정보학 툴을 사용하여 **리눅스 환경**에서 찾아보고자 함

* 미가공 데이터인 **FASTQ** 파일에서 reference 서열에 맞추어 **BAM** 파일을 만들고, 이 파일에서 유전적 변이를 찾아내는 Variant calling 과정을 거쳐 **VCF** 파일을 만든다

02_목표.jpg





**강의 목표**
* 리눅스 환경에서 DNA mutation(SNV Indel)을 검출하는 파이프라인 제작
* 파이프라인을 구성하는 각 툴을 설치하는 방법을 학습
* 각 툴들을 실행하여 분석 파이프라인을 완성하고 결과 파일 생성

**특징**
* BWA2, GATK4 같은 최신 툴 사용: raw data에서 변이 검출
* 최소한의 컴퓨팅 자원으로 변이 검출 파이프라인 제작 학습
* 로컬 PC에서 실습 진행 가능

**강의가 필요한 사람**
* DNA 유전체 분석의 파이프라인의 각 구성 요소를 배우고싶은 사람
* 바닥부터 툴을 하나씩 설치하며 유전체 분석 파이프라인을 만들어 보고 싶은 사람
* 최신의 분석 툴인 BWA2, GATK4로 유전체 분석을 진행하고 싶은 사람

**본 강의가 필요 없는 사람**
* 유전체 분석 전문가
* 유전체 분석 파이프라인의 세부적인 알고리즘을 알고 싶은 사람
